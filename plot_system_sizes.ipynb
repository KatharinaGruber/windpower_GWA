```python
import pandas as pd
import seaborn as sns
from refuelplot import *
setup()
sns.set_style("darkgrid")

from paths import results_path
```

```python
bra = pd.read_csv(results_path + '/BRA/number_grid_points.csv',index_col=0)
nz = pd.read_csv(results_path + '/NZ/number_grid_points.csv',index_col=0)
usa = pd.read_csv(results_path + '/USA/number_grid_points.csv',index_col=0)
zaf = pd.read_csv(results_path + '/ZAF/number_grid_points.csv',index_col=0)
bra['country'] = 'Brazil'
nz['country'] = 'New Zealand'
usa['country'] = 'USA'
zaf['country'] = 'South Africa'
```

```python
cols = ['country','dataset','cor']
ss = pd.concat([bra[cols],nz[cols],usa[cols],zaf[cols]]).rename({'cor':'system size'},axis=1)
```

```python
sns.catplot(x = 'dataset', y = 'system size', col='country',col_wrap=2,
              kind='violin',#hue='dataset',
            height=4, aspect=1,
            data=ss, margin_titles=True,sharey=False
           ).set_xticklabels(rotation=20)
```
